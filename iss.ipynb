```python
import requests
import time

latitudes = []
longitudes = []
timestamps = []


for i in range(100):
    try:
        response = requests.get('https://api.wheretheiss.at/v1/satellites/25544')
        response.raise_for_status()

        iss_location = response.json()

        latitude = iss_location['latitude']
        longitude = iss_location['longitude']
        timestamp = iss_location['timestamp']

        latitudes.append(latitude)
        longitudes.append(longitude)
        timestamps.append(timestamp)



        print(f"{i+1}:Latitude: {latitude}, Longitude: {longitude}, Timestamp: {timestamp}")

        time.sleep(10)

    except requests.exceptions.RequestException as e:
        print(f"An error occurred at point{i+1} {e}")
        time.sleep(10)


import matplotlib.pyplot as plt

plt.plot(longitudes, latitudes, marker='o')
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("ISS Path (100 Points)")
plt.grid(True)
plt.show()
```
